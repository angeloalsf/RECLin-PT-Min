## 5. Conferir os splits

Os splits (`data/splits/*.jsonl`) são versionados no repo, então já vêm com o clone. **São exatamente os mesmos usados pelo BioBERTpt** (seed 42, nível de documento). Se faltarem e os XMLs estiverem presentes, geramos na hora.

```python
import subprocess, sys
splits = os.path.join(REPO, 'data', 'splits')
need = ['train.jsonl', 'dev.jsonl', 'test.jsonl']
have = all(os.path.isfile(os.path.join(splits, f)) for f in need)

if not have:
    xml = os.path.join(REPO, 'SemClinBr-xml-public-v1')
    if os.path.isdir(xml) and os.listdir(xml):
        print('Splits ausentes; gerando a partir dos XMLs...')
        subprocess.run([sys.executable, 'src/parse_semclinbr.py',
                        '--xml-dir', 'SemClinBr-xml-public-v1',
                        '--out', 'data/processed/dataset.jsonl'], cwd=REPO, check=True)
        subprocess.run([sys.executable, 'src/make_splits.py'], cwd=REPO, check=True)
    else:
        raise FileNotFoundError(
            'Splits não encontrados e XMLs ausentes. Faça commit de data/splits/ '
            'no repo, ou suba a pasta SemClinBr-xml-public-v1/ para o Colab.')

for f in need:
    n = sum(1 for _ in open(os.path.join(splits, f), encoding='utf-8'))
    print(f'{f}: {n} docs')
```

## 6. Treinar o baseline BERTimbau (com retomada)

Encoder `neuralmind/bert-base-portuguese-cased`, marcadores de entidade `[E1] [/E1] [E2] [/E2]`, 3 classes, `class_weight=balanced`. Melhor época escolhida pelo macro-F1 no **dev**; **test** reportado uma única vez.

**Hiperparâmetros idênticos aos do BioBERTpt** (mesmo `--epochs/--batch-size/--max-gap/--max-length/--seed`) — só muda `--model`, `--ckpt-dir` e `--out`.

`--ckpt-dir` aponta para o Drive: ao fim de cada época grava `last_checkpoint/` (e `best_model/` quando bate o recorde no dev). Se o runtime cair, **rode esta célula de novo** (depois das de cima) e ele **pula as épocas já feitas**. Se já tiver feito todas, vai direto para a avaliação final.

```python
%cd $REPO
!python -u src/baseline_bertimbau.py \
    --splits-dir data/splits \
    --model neuralmind/bert-base-portuguese-cased \
    --epochs 3 \
    --batch-size 64 \
    --max-gap 20 \
    --max-length 128 \
    --seed 42 \
    --ckpt-dir "$CKPT_DIR" \
    --out results/baseline_bertimbau.json
```

## 7. Ver os resultados (métricas + matriz de confusão)

```python
import json
r = json.load(open(os.path.join(REPO, 'results', 'baseline_bertimbau.json'), encoding='utf-8'))
print('Modelo:', r['model'])
print('Candidatos:', r['n_candidates'])
print('Macro-F1:', round(r['test_macro_f1'], 4))
print('Micro-F1:', round(r['test_micro_f1'], 4))
print('\nF1 por classe:')
for k, v in r['test_f1_per_class'].items():
    mark = '   <<< NEGAÇÃO' if k == 'negation_of' else ''
    print(f'  {k:<18s} {v:.4f}{mark}')

cm = r['confusion_matrix']
labels = cm['labels']
print('\nMatriz de confusão (linhas=verdadeiro, colunas=predito):')
short = [l[:8] for l in labels]
print('true\\pred'.ljust(12) + ''.join(s.rjust(10) for s in short))
for i, l in enumerate(labels):
    print(l[:11].ljust(12) + ''.join(str(int(c)).rjust(10) for c in cm['matrix'][i]))
```
